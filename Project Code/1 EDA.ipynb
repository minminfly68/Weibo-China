```python
import operator
df = pd.read_csv("2656274875.csv")
ls_key = []
for row in df.iterrows():
    s = row[1][2]
    for keyword, weight in textrank(s, withWeight=True, topK=1):
        temp_key = keyword
    if not temp_key.isdigit():
        ls_key.append(temp_key)

dict_key = dict((x,ls_key.count(x)) for x in set(ls_key))
dict_key = dict(sorted(dict_key.items(), key=operator.itemgetter(1),reverse=True))
df_output = pd.DataFrame(list(dict_key.items()), columns=['topic', 'frequency'])
df_output.to_csv("2 topic_extraction_output_rank.csv", index=False, encoding='utf_8_sig')
```

```python
df = pd.read_csv("1699432410.csv")
df.head()
```

```python
ls_keyword = ['民主','自由','法治','人权','政治','干涉','内政','抗议', '中方', '美方']
ls_temp = []
for row in df.itertuples():
    identifier = False
    for word in ls_keyword:
        if word in row[3]:
            ls_temp.append(True)
            identifier = True
            break
    if not identifier:
        ls_temp.append(False)

df['politics'] = ls_temp
df['politics'].value_counts()
df.to_csv('output.csv', index=False, encoding='utf_8_sig')
```
